```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
observations= 1000
x= np.random.uniform(low=-10, high=10, size=(observations,1))
z= np.random.uniform(low=-10, high=10, size=(observations,1))
inputs= np.column_stack((x,z))
noise= np.random.uniform(low=-1, high=1, size= (observations,1))
target= 2*x-3*z+5+noise
np.savez('TF-First', inputs= inputs, targets= target)
```

```python
training_data= np.load('TF-First.npz')
```

```python
input_size=2
output_size=1
```

```python
model= tf.keras.Sequential([
           tf.keras.layers.Dense(output_size)
])
```

```python
model.compile(optimizer= 'sgd', loss= 'mean_squared_error')
```

```python
model.fit(training_data['inputs'], training_data['targets'],epochs=10, verbose=2)
```

```python
model.layers[0].get_weights()
```

```python
model.predict_on_batch(training_data['inputs']).round(1)
```

```python
training_data['targets'].round(1)
```

```python
plt.plot(np.squeeze(model.predict_on_batch(training_data['inputs'])), np.squeeze(training_data['targets']))
plt.xlabel('outputs')
plt.ylabel('targets')
plt.show()
```
